```python
# set up and load data, checking we've gotten it all
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import string
```

```python
mpesa_started=2007
```

```python
path="C:\\Users\\Nicholas\\text Analysis\\MPESA_Statement.csv"
attributes=["Transaction_code","Transaction_date"]
df=pd.read_csv(path,names=attributes)
display(df.head(n=1000))
```

```python
df['Convert_Date'] = pd.to_datetime(df['Transaction_date'])
df['Converted_Date']= df['Convert_Date'].dt.strftime('%Y-%m-%d')
df
```

```python
df['Final_trans_date'] = df['Convert_Date'].dt.normalize()
df['Final_transaction_date']= df['Final_trans_date'].dt.strftime('%Y-%m-%d')
df
```

```python
df['Decoded_Year'] = df['Transaction_code'].str[:1].str.lower()

df['y-m'] = df['Transaction_code'].str[:2]
df['Decoded_Month'] = df['y-m'].str[-1].str.lower()

df['y-m-d'] = df['Transaction_code'].str[:3]
df['Decoded_Date'] = df['y-m-d'].str[-1].str.lower()
df
```

```python
def mpesa_y_code(text):

    dict = {'a':'1','b':'2','c':'3','d':'4','e':'5','f':'6','g':'7','h':'8','i':'9','j':'10','k':'11','l':'12','m':'13','n':'14','o':'15','p':'16','q':'17','r':'18','s':'19','t':'20','u':'21','v':'22','w':'23','x':'24','y':'25','z':'26'}
    new_text = text
    for i in new_text:
        if i in dict:
            new_text = new_text.replace(i, dict[i])
    return new_text
```

```python
df['Year'] = df["Decoded_Year"].apply(mpesa_y_code)
```

```python
def mpesa_m_code(text):

    dict = {'a':'01','b':'02','c':'03','d':'04','e':'05','f':'06','g':'07','h':'08','i':'09','j':'10','k':'11','l':'12','m':'13','n':'14','o':'15','p':'16','q':'17','r':'18','s':'19','t':'20','u':'21','v':'22','w':'23','x':'24','y':'25','z':'26'}
    new_text = text
    for i in new_text:
        if i in dict:
            new_text = new_text.replace(i, dict[i])
    return new_text
```

```python
df['Month']= df['Decoded_Month'].apply(mpesa_m_code)
```

```python
def mpesa_d_code(text):

    dict = {'1':'01','2':'02','3':'03','4':'04','5':'05','6':'06','7':'07','8':'08','9':'09', 'a':'10','b':'11','c':'12','d':'13','e':'14','f':'15','g':'16','h':'17','i':'18','j':'19','k':'20','l':'21','m':'22','n':'23','o':'24','p':'25','q':'26','r':'27','s':'28','t':'29','u':'30','v':'31'}
    new_text = text
    for i in new_text:
        if i in dict:
            new_text = new_text.replace(i, dict[i])
    return new_text
```

```python
df['date']=df['Decoded_Date'].apply(mpesa_d_code)
```

```python
df.insert(loc=7, column='Mpesa_Started', value=mpesa_started)
df
```

```python
df["Trans_Year"] = (df["Mpesa_Started"] + df["Year"].astype(int))-2
df
```
